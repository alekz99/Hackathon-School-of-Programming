```python
def get_managers_replicas(row):
  manager_id = int(row['manager_id'])
  if manager_id == -1:
    # менеджер не найден, добавляем весь звонок
    sentences = []
    for text_row in str(row['texts']).split('\n'):
      if 'SPEAKER' in text_row:
        sentences.append(text_row[12:])
  else:
    sentences = []
    for text_row in str(row['texts']).split('\n'):
      if f'SPEAKER_0{manager_id}' in text_row:
        sentences.append(text_row[12:])
  return ' '.join(sentences)

data['managers_replicas'] = data.apply(get_managers_replicas, axis=1)
```

```python
def get_clients_replicas(row):
  manager_id = int(row['manager_id'])
  if manager_id == -1:
    # менеджер не найден, добавляем весь звонок
    sentences = []
    for text_row in str(row['texts']).split('\n'):
      if 'SPEAKER' in text_row:
        sentences.append(text_row[12:])
  else:
    sentences = []
    for text_row in str(row['texts']).split('\n'):
      if f'SPEAKER_0{manager_id}' not in text_row:
        sentences.append(text_row[12:])
  return ' '.join(sentences)

data['clients_replicas'] = data.apply(get_clients_replicas, axis=1)
```

```python
import torch
import numpy as np

from tqdm import tqdm
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from nltk.tokenize import sent_tokenize
```

```python
model_checkpoint = 'cointegrated/rubert-tiny-sentiment-balanced'
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)
if torch.cuda.is_available():
    model.cuda()

def get_sentiment(text):
    """ Calculate sentiment of a text. `return_type` can be 'label', 'score' or 'proba' """
    with torch.no_grad():
        inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True).to(model.device)
        proba = torch.sigmoid(model(**inputs).logits).cpu().numpy()[0]
    # label, score, proba
    # 'negative', -0.5894946306943893, [0.7870447  0.4947824  0.19755007]
    return model.config.id2label[proba.argmax()], proba.dot([-1, 0, 1]), proba
```

```python
score_clinets = []
filter_del_auto_reply = (data['one_spicker'] == 1) & (data['manager_id'] == -1)
data_clients = data[~filter_del_auto_reply]

for index, row in tqdm(data_clients.iterrows()):
  score_client = []
  # УДАЛЯЕМ ДУБЛИ (заикания модели транскрибации)
  sentences = set(sent_tokenize(row['clients_replicas']))
  for sentence in sentences:
    label, score, proba = get_sentiment(sentence)
    score_client.append(score)
  score_clinets.append(score_client)
```

```python
result_score_clinets = []
max_score_clinets = []
min_score_clinets = []

for s in score_clinets:
  if len(s) != 0:
    #temp = [np.sign(numb) * (numb ** 2) for numb in s]
    #sum_temp = sum(temp)
    #result_score_clinets.append(np.sign(sum_temp) * ((np.sign(sum_temp) * sum_temp) ** 0.5) / len(temp))
    result_score_clinets.append(sum(s) / len(s))
    max_score_clinets.append(max(s))
    min_score_clinets.append(min(s))
  else:
    result_score_clinets.append(0)
    max_score_clinets.append(0)
    min_score_clinets.append(0)

data_clients['avg_clients_score'] = result_score_clinets
data_clients['max_clients_score'] = max_score_clinets
data_clients['min_clients_score'] = min_score_clinets
```

## Подготовка выходных данных

Алгоритм сборки данных:

1.   Звонки, которые определены как СОСТОЯВШИЕСЯ, группируются и выставляется среднее для всех звонков. Так получаем классификацию общения с клиентом.
2.   Из группы НЕСОСТОЯВШИХСЯ звонков отбираем звонки, которые были клиентам, которых нет в первой группе. Группируем их по клиентам, и получаем, что для этих клиентов общение не состоялось.



```python
data_clients['user_id'] = data_clients['names'].apply(lambda text: text.split('_')[0])
data_clients['call_date'] = data_clients['names'].apply(lambda text: pd.to_datetime(f"{text.split('_')[1]} {text.split('_')[2].replace('.', ':')[:8]}"))
```

```python
def calculate_score(avg_clients_scores):
  score = sum(avg_clients_scores) / len(avg_clients_scores)
  if score < -0.2:
    return 1
  elif score >= 0.2:
    return 3
  else:
    return 2

group_data_clients = data_clients.sort_values(['user_id', 'call_date'], ascending = True).groupby(['user_id'])['avg_clients_score'].apply(calculate_score).reset_index()
group_data_clients.rename(columns={"avg_clients_score": "class"}, inplace=True)
group_data_clients['class'].value_counts()
```

```python
data['user_id'] = data['names'].apply(lambda text: text.split('_')[0])
data['call_date'] = data['names'].apply(lambda text: pd.to_datetime(f"{text.split('_')[1]} {text.split('_')[2].replace('.', ':')[:8]}"))

already_clients = group_data_clients.user_id.to_list()
filter_already_clients = ~data['user_id'].isin(already_clients)
```

```python
group_data_auto_reply = data[filter_del_auto_reply & filter_already_clients].sort_values(['user_id', 'call_date'], ascending = True).groupby(['user_id'])['texts'].apply(lambda x: 0).reset_index()
group_data_auto_reply.rename(columns={"texts": "class"}, inplace=True)
group_data_auto_reply['class'].value_counts()
```

```python
class_transform = {
    0: 'общение не состоялось',
    1: 'негативный клиент',
    2: 'нейтральный клиент',
    3: 'заинтересованный клиент'
}

columns = {"user_id": "lead_id", "class": "class_type"}
output = pd.concat([group_data_clients, group_data_auto_reply], ignore_index=True).sort_values(['user_id'], ascending=True).rename(columns=columns)
output['class_type'] = output['class_type'].apply(lambda x: class_transform[x])
output
```
